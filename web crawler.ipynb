```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, StaleElementReferenceException
import time
import pandas as pd
from elasticsearch import Elasticsearch
import re
```

```python
driver = webdriver.Chrome()
```

```python
maxpage = 10
data =[]
for i in range(1,maxpage+1):
    #get webpages one by one
    url = f"https://messari.io/assets?page={i}"
    driver.get(url)
    #just another wait for certeinity :))
    driver.implicitly_wait(15)
    time.sleep(5) #let the web loads
    #close the popup if it appears
    if i==1:
        try:
            popup_window = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.CLASS_NAME, "MuiDialog-paper"))
            )

            # Locate and click the close button
            close_button = WebDriverWait(driver, 20).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "button[title='Close dialog']"))
            )
            close_button.click()
        except NoSuchElementException,TimeoutException:
            print("this time there was no popup")
        
    table = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//table[@class='MesDataTable-table mui-a8722g']")))
    # Get all rows from the table
    rows = table.find_elements(By.TAG_NAME, "tr")


    
    # Iterate over rows and extract data
    for row in rows:
        driver.implicitly_wait(5)
        # Get all cells in the row
        #cells = WebDriverWait(driver, 0.5,ignored_exceptions=(NoSuchElementException, StaleElementReferenceException)).until(EC.presence_of_all_elements_located((By.TAG_NAME, "tr")))
        cells= table.find_elements(By.TAG_NAME, "td")
        # Extract data from each cell
        row_data = [cell.text for cell in cells]
        if row_data==[]:continue
        # Append the row data to the list
        data.append(row_data)

# Create a DataFrame from the list of data
full_data = pd.DataFrame(data,columns=['Name','Price_current','Price_24H','Price_7D','Price_30D','Price_30DTL','Marketcap','Real_Volume','Sector'])



# Print the DataFrame
full_data
```

```python
#this function edit coin names beacause they have \n at first of them
def edit_name(name):
    
    edited_name = name.split('\n')[1]  # Retain only the part after the first '\n'
    return edited_name
```

```python
filtered_data = full_data[full_data['Name'].notna()] # removes nans
#filtered_data = filtered_data.drop(columns="Price_30DTL") #remove coloumn
filtered_data["Name"] = filtered_data["Name"].apply(edit_name) # edit names

filtered_data
```

```python
#this function is for edit index names
def sanitize_index_name(name):
    # Replace spaces with underscores and convert to lowercase
    name = name.replace(" ", "_").lower()
    # Remove invalid characters
    name = re.sub(r'[^a-z0-9_]', '', name)
    return name
```

```python
def extract_numeric_value(s):
    match = re.search(r'[-+]?\d*\.\d+|\d+', s)
    if match:
        return float(match.group())
    else:
        return None
```

```python
sorted_data_30D = pd.DataFrame(columns=['Name','Price_current','Price_24H','Price_7D','Price_30D','Marketcap','Real_Volume','Sector'])
#sort data by their 30 Days price change base on their sector
filtered_data_copy = filtered_data.copy()
filtered_data_copy['Price_30D'] = filtered_data_copy['Price_30D'].apply(extract_numeric_value)

sector_percentages = {}

grouped = filtered_data_copy.groupby("Sector")
for sector, group in grouped:
    # Sort data within each group by the desired column
    sorted_data_30D = pd.concat([sorted_data_30D,group.sort_values(by="Price_current",ascending=False)] ,ignore_index=True)

    ascending_assets = group[group['Price_30D'] > 0]
    percent_ascending = (len(ascending_assets) / len(group)) * 100
    sector_percentages[sector] = percent_ascending

# Print the percentages for each sector
for sector, percentage in sector_percentages.items():
    print(f"Percentage of assets with ascending trend in {sector}: {percentage:.2f}%")
    
    
sorted_data_30D
```

```python
def extract_numeric_value_volume(s):
    match = re.search(r'[-+]?\d*\.\d+|\d+', s)
    if match:
        volume = float(match.group())
        if s[-1]=='B':
            return volume * 10 **9 
        elif s[-1]=='M':
            return volume * 10 **6 
        elif s[-1]=='K':
            return volume * 10 **3 
        
    else:
        return None
```

```python
edited_data = pd.DataFrame(columns=['Name','Price_current','Price_24H','Price_7D','Price_30D','Marketcap','Real_Volume','Sector'])
#sort data by their volumes price change base on their sector
filtered_data_copy1 = filtered_data.copy()
filtered_data_copy1['Real_Volume'] = filtered_data_copy['Real_Volume'].apply(extract_numeric_value_volume)

means = dict()
grouped1 = filtered_data_copy1.groupby("Sector")
for sector, group in grouped1:
    means[sector] = group['Real_Volume'].mean()
    edited_data = pd.concat([edited_data,group.sort_values(by="Real_Volume",ascending=False)] ,ignore_index=True)
 
means = pd.DataFrame.from_dict(means, orient='index', columns=['Mean_Volume'])
means
```
